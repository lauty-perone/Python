<h3>1)Primero que todo y lo principal seria descargar los archivos.txt , posteriormente abrirlos en modo de lectura          para poder almacenarlo en una lista. Y un detalle no menos importante es que al abrir el archivo se decodifica de binario a string.
<h3>Detalle :importo la funci√≥n math para luego usar un m√©todo que me permitira truncar un n√∫mero

```python
import math

arch_nombres = open("nombres.txt", "r", encoding="utf-8")
nombres = arch_nombres.read().split()  
arch_nombres.close()  

arch_notas_1= open('eval1.txt','r',encoding="utf-8")
notas_1 = map (int, arch_notas_1.read().split(","))
arch_notas_1.close()

arch_notas_2 = open('eval2.txt','r',encoding="utf-8")
notas_2 = map(int,arch_notas_2.read().split(","))  
arch_notas_2.close()
```

<h3>2)Uso un for para reemplazar el caracter ,(coma) que queda en la lista nombres, para que quede m√°s est√©tico 

```python
for pos, nombre in enumerate(nombres):
    nombres[pos] = ((nombre. replace(",",''))) 
```

<h3>3) Posteriormente a leer los archivos de las notas, creo una lista con la suma de ambas listas en sus respectivas posiciones, utilizando la expresi√≥n lambda para ir sumando y luego haciendo un map para que me retorne una lista con el resultado de lambda. 

```python
total = list(map(lambda x,y: x + y, notas_1,notas_2)) 
```

<h3>4) Una vez hecha la lista con el total de ambas notas, creo un diccionario con la instrucci√≥n zip que me pone el primer par√°metro("nombres") como las claves y el segundo ("total") como los valores y por supuesto en sus lugares adecuados.

```python
estructura_dicci = dict(zip(nombres, total)) 
```

<h3>5)A continuaci√≥n calculo el promedio con la suma de la lista "total" y la divido por su cantidad de elementos resultando as√≠ un n√∫mero con coma al que trunque para que me quede en entero con el m√©todo trunc de la funci√≥n math importada anteriormente.
   <h3> Luego imprimo el promedio.

```python
prom=  math.trunc(sum(total) /len(total)) 
print('El promedio total de alumnos es: ' ,prom)
```

<h3>6)Luego de calcular el promedio recorro el diccionario preguntando si la suma de sus dos notas del alumno ,de esa posici√≥n, es mayor al promedio. 
    <h3>Si es asi, imprimo su nombre.

```python
for elem in estructura_dicci:
    if estructura_dicci[elem]> prom: 
        print('El nombre del alumno que supera el promedio es: ',elem)
```
